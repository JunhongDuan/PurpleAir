```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
pm25_2020 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm25_2020.csv')
pm25_2021 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm25_2021.csv')
pm25_2022 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm25_2022.csv')

pm10_2020 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm10_2020.csv')
pm10_2021 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm10_2021.csv')
pm10_2022 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm10_2022.csv')

no2_2020 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_no2_2020.csv')
no2_2021 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_no2_2021.csv')
no2_2022 = pd.read_csv('../Data/Raw data/EPA air quality data/ad_viz_plotval_data_no2_2022.csv')
pm25_2020.head()
```

```python
# merge all year data and cut data within 2020-12-1 to 2022-12-1
paths = [
    "../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm25_2020.csv",
    "../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm25_2021.csv",
    "../Data/Raw data/EPA air quality data/ad_viz_plotval_data_pm25_2022.csv"
]
df_list = [pd.read_csv(p) for p in paths]
df_pm25 = pd.concat(df_list, ignore_index=True)

# filter date
df_pm25['Date'] = pd.to_datetime(df_pm25['Date'])
start = pd.Timestamp("2020-12-01")
end   = pd.Timestamp("2022-12-01")
mask = (df_pm25['Date'] >= start) & (df_pm25['Date'] <= end)
df_pm25_filtered = df_pm25.loc[mask].copy()
```

```python
# covert into shapefile
df_pm25_filtered = gpd.GeoDataFrame(
    df_pm25_filtered,
    geometry=gpd.points_from_xy(df_pm25_filtered['Site Longitude'], df_pm25_filtered['Site Latitude']),
    crs="EPSG:4326"
)
df_pm25_filtered['Date'] = df_pm25_filtered['Date'].dt.date.astype(str)
df_pm25_filtered.to_file('../Data/Output/epa_pm25_history.shp')
```
